# Kodeforståelse

**Læringsmål:**

* Tolke kode

Koden under kan ikke kjøres, her skal du tenke deg fram til svarene! Ca. 20% av eksamen er kodeforståelse så se på dette som god trening! 

**a)** Hva skrives ut i koden under?

```python
a=345
b=''
while a or b=='':
    b=str(a%2)+b
    a=a//2
print(b)
```

Svar: 101011001 (Hver modulo blir lagt til på høyre siden av stringen til a blir lik 0==False)

**b)** Hva skrives ut i koden under?

```python
for x in range(0, 10, 2):
    print(x, end='')
    if x%4==0:
        print( ": Dette tallet går opp i 4-gangern")
    else:
        print()
     ```
     
`end=""` gjør at det neste som printes ikke printes en linje under, men at det fortsetter på samme linje.

Svar: (Rangen ekskluderer 10 og har steps på 2)
      0: Dette tallet går opp i 4-gangern 
      2
      4: Dette tallet går opp i 4-gangern
      6
      8: Dette tallet går opp i 4-gangern

**c)** Hva skrives ut i koden under?

```python
i = 1
while i<10:
    i = i*2
print(i)
```

Svar: 16 (Printen skjer utenfor while loopen)
      

**d)** Hva skrives ut i koden under?

```python
i = 1
j = 3
while j>0:
    i = i*2
    j = j - 1
print(i)```

Svar: 8 (j blir 0 etter i bli 8)

**e)** Hva skrives ut i koden under?

```python
i = 5
for x in range(i):
    for y in range(x+1):
        print("*", end="")
    print()```
    
Her er det en dobbel løkke, så dette er nok nytt for mange. Prøv likevel! Hvordan fungerer egentlig en løkke i en løkke? Se side 176 i Starting Out with Python.

Svar: (Den første for-løkken bestemmer at det blir 5 linjer, den andre for-løkken bestemmer hvor mange ganger "*" printes etter hverandre i hver linje) (Når range er på 5 vil x starte som 0, for å få en stjerne første gang må vi skrive x+1, slik at den første rangen blir 1 og at vi faktisk får en stjerne printet ut)
      *
      **
      ***
      ****
      *****
